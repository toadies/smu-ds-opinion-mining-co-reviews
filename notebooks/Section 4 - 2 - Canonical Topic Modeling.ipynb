```python
Integrity = ['integrity', 'ethics', 'ethic' , 'accountability' , 'trust' , 'honesty' , 'responsibility' , 'fairness' , 'do the right thing' , 'transparency' , 'ownership']
Teamwork = ['teamwork','collaboration','cooperation']
Innovation = ['innovation','creativity','excellence','improvement','passion','pride','leadership','growth','performance','efficiency','results']
Respect = ['respect','diversity','inclusion','development','talent','employees','dignity','empowerment']
Quality = ['quality','customer','meet','needs','commitment','make a difference','dedication','value','exceed expectations']
Safety = ['safety','health','work life balance', 'flexibility']
Community = ['community','environment','caring','citizenship']
Communication = ['communication','openness']
```

```python
def find_topic(review, topics):
    found_topic = []
    tokins = review.split(" ")
    for tokin in tokins:
        if tokin in topics:
            found_topic.append(tokin)
    return ', '.join(found_topic)
```

```python
df = pd.DataFrame(tech_review_corpus)

df['Integrity'] = df.review.apply(find_topic,topics=Integrity)
df['Teamwork'] = df.review.apply(find_topic,topics=Teamwork)
df['Innovation'] = df.review.apply(find_topic,topics=Innovation)
df['Quality'] = df.review.apply(find_topic,topics=Quality)
df['Safety'] = df.review.apply(find_topic,topics=Safety)
df['Community'] = df.review.apply(find_topic,topics=Community)
df['Communication'] = df.review.apply(find_topic,topics=Communication)
```
