```python
import pandas as pd
import pickle

with open("../data/all_reviews.pkl","rb") as f:
    reviews = pickle.load(f)
```

```python
reviews.drop_duplicates(inplace=True)

reviews["job_title"].dropna(inplace=True)
```

```python
reviews["job_title"].isna().sum()
```

```python
job_titles = reviews.job_title.tolist()

job_titles = list( set(job_titles) )
print("Total Job Titles", len(job_titles))
print(job_titles[:6])
```

```python
job_list=[
    "software developer"
    ,"web developer"
    ,"applications"
    ,"system administrator"
    ,"computer programmer"
    ,"DevOps"
    ,"cloud"
    ,"Database"
    ,"agile"
    ,"system"
    ,"technology"
    ,"iOS"
    ,"Java"
    ,"analytics"
    ,"scientist "
]
```

```python
from collections import defaultdict
from gensim import corpora
```

```python
# remove common words and tokenize
stoplist = set('for a of the and to in'.split())
texts = [
    [word for word in title.lower().split() if word not in stoplist]
    for title in job_list
]
print(texts)
# remove words that appear only once

dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
print(dictionary)
```

```python
from gensim import models
lsi = models.LsiModel(corpus, id2word=dictionary, num_topics=2)
```

```python
vec_lsi=[]
doc = job_titles
for item in doc:
    vec_bow = dictionary.doc2bow(item.lower().split())
    vec_lsi.append(lsi[vec_bow]) # convert the query to LSI space
    
   
```

```python
from gensim import similarities
index = similarities.MatrixSimilarity(lsi[corpus])  # transform corpus to LSI space and index it
```

```python
sims = index[vec_lsi]  # perform a similarity query against the corpus
```

```python
len(sims)
```

```python
#sims = sorted(enumerate(sims), key=lambda item: -item[1])
TechJobTitel = []
for i in range(len(sims)):
    if sims[i].mean() > 0:
        SimMean = sims[i].mean()
        print(sims[i].mean(),doc[i])
        TechJobTitel.append(doc[i])
  
```
